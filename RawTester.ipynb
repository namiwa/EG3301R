```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
DIRECTORY = os.getcwd()
FILEDIR = os.path.join(DIRECTORY, "Jafarabad coastal mast data")
FILE = os.path.join(FILEDIR, "Jafarabad coastal mast data.txt")
COLS = ["Date/Time", "100m_NWS[m/s]", "100m_N", "Std[m/s]", "100m_SWS[m/s]", "100m_Std[m/s]", "80m_WS[m/s]", "80m_Std[m/s]", "50m_WS[m/s]", "50m_Std[m/s]",  "20m_WS[m/s]", "20m_Std[m/s]", "Pressure[m]", "98m_DIR[]",  "78m_DIR[]",  "48m_DIR[]"]

data = pd.read_csv(FILE, sep="\t")
data.columns = COLS
data.head()
```

```python
data.dtypes
```

```python
datetime_series = pd.to_datetime(data['Date/Time'])
datetime_index = pd.DatetimeIndex(datetime_series.values)
```

```python
data.index = datetime_index
```

```python
data.drop('Date/Time', axis=1, inplace=True)
data.head()
```

```python
lines = data.plot.line(subplots=True, figsize=(40,20))
```
